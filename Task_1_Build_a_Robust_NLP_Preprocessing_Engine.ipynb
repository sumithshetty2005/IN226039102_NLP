# Task 1: Conceptual Understanding

**1) What is the difference between "Love" and "love" in NLP?** <br>
In raw NLP, these are treated as two distinct tokens because computers use ASCII/Unicode values to identify characters. " L" (65) and "l" (108) are different. If you don't lowercase your text, your model might think "Love" at the start of a sentence is a different concept than "love" in the middle of one, splitting your data's statistical power.
<br>
<br>
**2) What happens if stopwords are not removed?** <br>
When you're looking at the words that show up a lot, like "the", "is", and "at", they tend to be all over the place in every document. The problem is, these words don't really help us figure out what's unique about a particular topic, because they're used just as much in a sports article as they are in a cooking blog. For example, both types of articles use the word "the" all the time. This can cause a few issues, like making our models, such as Naive Bayes, more expensive to run and less accurate. We need to find a way to focus on the words that really matter, the ones that will help us tell the difference between a sports article and a cooking blog. By getting rid of these common words, we can make our models work better and faster.
<br>
<br>
**3) Provide two real-world scenarios where removing stopwords can be harmful**<br>
Sentiment Analysis: Removing "not" from "I am not happy" turns the sentiment into "happy," which is a catastrophic error.
Search Queries / Specific Phrases: In a phrase like "To be or not to be," removing stopwords leaves you with... nothing. The meaning is entirely contained within the small words.
<br>
<br>
**4) Explain the difference between stemming and lemmatization.**<br>
Stemming is a pretty basic way of shortening words. It just cuts off the endings, like when you change "running" to "run". But sometimes it doesn't work so well, like with the word "studies" - it might get shortened to "studi", which isn't even a real word. It's a fast way to do things, but it's not always accurate.

Lemmatization: A sophisticated "dictionary-based" approach. It looks at the root word (lemma) based on the word's part of speech (e.g., "better" becomes "good"). It is more accurate but slower.


# Task 2: Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return []

    text = text.lower()
    text = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', ' ', text)
    raw_tokens = text.split()
    meaningful_short_words = {'no', 'not'}
    cleaned_tokens = [t for t in raw_tokens if len(t) > 2 or t in meaningful_short_words]

    return cleaned_tokens
```

# Task 3: Stress Testing

```python
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

processed_results = []

for original in sample_inputs:
    tokens = preprocess_text(original)
    cleaned_sentence = " ".join(tokens)
    processed_results.append({
        "original": original,
        "tokens": tokens,
        "cleaned": cleaned_sentence
    })

    print(f"Original: {original}")
    print(f"Tokens:   {tokens}")
    print(f"Cleaned:  {cleaned_sentence}")
    print("-" * 30)
```

# Task 4: Token Analytics

```python
for item in processed_results:
    tokens = item['tokens']
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total_tokens if total_tokens > 0 else 0

    print(f"Sentence: {item['cleaned']}")
    print(f"Total: {total_tokens} | Unique: {unique_tokens} | Avg Length: {avg_len:.2f}\n")
```

```python
# --- Analysis Answers ---
# 1. Most Noise: "Visit https://openai.com now!" (URL and punctuation removed)
#    and "Call me at 9876543210" (Numbers and small words removed).
# 2. Most Meaningful: "I am not happy with this" (Retains "not" and "happy")
#    and "I absolutely looooved this product" (Retains key descriptors).
```

# Task 5: Task Analysis

```python
from collections import Counter

all_tokens = [token for item in processed_results for token in item['tokens']]
word_counts = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(word_counts.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(word_counts.most_common()[:-6:-1])
```

# Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    clean_tokens_flat = []
    clean_sentences_list = []

    for text in text_list:
        tokens = preprocess_text(text)
        clean_tokens_flat.extend(tokens)
        clean_sentences_list.append(" ".join(tokens))

    return {
        "tokens": clean_tokens_flat,
        "clean_sentences": clean_sentences_list
    }

pipeline_output = full_pipeline(sample_inputs)
print(pipeline_output)
```
